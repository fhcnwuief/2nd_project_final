### notedata의 season을 sdf 데이터 프레임의 pk에 맞는 방식으로 수정

1. 해당 데이터프레임을 리스트로 변경
2. 0>1 1>2 2>3 3>4 로 변경
3. 리스트를 문자열로 변경
4. 새로운 데이터프레임으로 만들고 기존 데이터에 존재하는 값 삭제, 후 새로운 데이터프레임 붙이기

```python
season = notedata['season']
season
```

```python
season_list = notedata['season'].to_list()
type(season_list)
```

```python
big_list=[]
for num in season_list:
    new_list = []
    for s in range(len(num)):
        if num[s].isnumeric():
            n = int(num[s])+1
            new_list.append(n)
    new_list
    big_list.append(new_list)
print(big_list)
```

```python
seadf1 = pd.DataFrame([big_list])
seadf1 = seadf1.T
seadf1.dtypes
```

```python
seadf1.rename(columns = {0 : 'season'}, inplace = True)
seadf1.columns
```

```python
seadf1['season'] = seadf1['season'].astype(str)
seadf1
```

```python
seadf1[seadf1['season']=='[1, 2, 3, 4]']
```
