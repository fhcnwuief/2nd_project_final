```python
import pandas as pd
```

```python
df = pd.read_csv('../data/middle_count.csv', index_col = 0)
df.sort_values(by = 'count', ascending = False)
note_list = df.nlargest(10, 'count')['middle_note']
```

```python
df.sort_values(by = 'count',ascending = False)
```

```python
note_df = pd.read_csv('../data/all_note_data.csv')
note_df
```

```python
cat = []
for i in note_list:
    for j in range(len(note_df)):
        if note_df['note'][j] == i:
            cat.append(note_df['category_name'][j])
cat
```

```python
cat_idx = []
for i in note_list:
    for j in range(len(note_df)):
        if note_df['note'][j] == i:
            cat_idx.append(note_df['category_idx'][j])
cat_idx
```
