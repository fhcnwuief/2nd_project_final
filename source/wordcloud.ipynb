```python
import pickle
import pandas as pd
import numpy as np
from PIL import Image
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
with open('../data/top_count.pickle', 'rb') as f:
    top_counts_dict = pickle.load(f)
    
perfume_mask = np.array(Image.open('../data/ (1).png'))
wordcloud = WordCloud(width=800, height=800,
                      background_color='white',
                      min_font_size=30, mask=perfume_mask).generate_from_frequencies(top_counts_dict)

plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
with open('../data/middle_count.pickle', 'rb') as f:
    top_counts_dict = pickle.load(f)
    
perfume_mask = np.array(Image.open('../data/ (1).png'))
wordcloud = WordCloud(width=800, height=800,
                      background_color='white',
                      min_font_size=25, mask=perfume_mask, colormap = 'gist_heat').generate_from_frequencies(top_counts_dict)

plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
