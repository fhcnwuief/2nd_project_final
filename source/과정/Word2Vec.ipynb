```python
# Compute similarity matrix. Higher score indicates greater similarity.
sim = []
for i in range(len(name2)):
    for j in range(i,len(name2)):
        sim.append(["(" + name2[i] + "," + name2[j] + ")", np.inner(embeddings[i], embeddings[j])])
sim
```

```python
import pandas as pd
data = pd.read_csv('../data/concat_category.csv')
data['name'].to_csv('../data/list.txt',index=False)

embed = hub.load("https://tfhub.dev/google/Wiki-words-250-with-normalization/2")
```

```python
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence

sentences = LineSentence("../data/list.txt")  # 전처리된 코퍼스를 로드합니다.
model = Word2Vec(sentences, vector_size=300, window=5, min_count=10, sg=1)  # Skip-gram 모델
# model = Word2Vec(sentences, vector_size=300, window=5, min_count=5, sg=0)  # CBOW 모델

model.save('../data/word2vec_model.bin')
```

```python
from gensim.models import KeyedVectors
import numpy as np
from scipy.spatial.distance import cosine

def cosine_similarity(embedding1, embedding2):
    return 1 - cosine(embedding1, embedding2)

# 사전 학습된 word2vec 모델을 불러옵니다.
model_path = '../data/word2vec_model.bin'  # 사전 학습된 모델의 경로로 대체해주세요.
word2vec_model = KeyedVectors.load(model_path)

def get_word_embedding(word):
    # 단어가 word2vec 모델의 어휘에 있는지 확인합니다.
    if word in word2vec_model.wv:
        return word2vec_model.wv[word]
    else:
        # 단어가 어휘에 없는 경우 처리
        return None

# 예시 단어
word = data['name'].to_list()
embedding_list = []
word_embedding_list = {}
# 단어의 임베딩 벡터를 가져옵니다.
for words in word:
     
    embedding = get_word_embedding(words.lower())  # 일관성을 위해 단어를 모두 소문자로 변환합니다.
    

    if embedding is not None:
        #print(f"'{words}'의 임베딩 벡터: {embedding}")
        word_embedding_list[words] = np.array(embedding)
        
    else:
        print(f"'{words}'은(는) 어휘에 없습니다.")

# 비교할 예시 단어들
word1 = 'flowers'
word2 = 'white_flowers'

if word1 in word_embedding_list and word2 in word_embedding_list:
    similarity = cosine_similarity(word_embedding_list[word1], word_embedding_list[word2])
    print(f"'{word1}'와(과) '{word2}'의 코사인 유사도: {similarity:.4f}")
else:
    print("두 단어 모두에 대한 단어 임베딩이 제공되지 않았습니다.")
```

```python
sim = {}
for i in word:
    sim[i] = {}
    for j in word:
        sim[i][j] = cosine_similarity(word_embedding_list[i], word_embedding_list[j])
sim
```

```python
test_df = pd.DataFrame.from_dict(sim)
test_df.to_csv('../data/cosine_sim.csv')
df = pd.read_csv('../data/cosine_sim.csv', index_col = 0)
df
```

각 카테고리별 top 5

```python
df = pd.read_csv('../data/cosine_sim.csv', index_col = 0)
abs_df = df.abs()

col_n = ['1', '2', '3', '4', '5']
index_n = abs_df.index.to_list()
top5 = pd.DataFrame(columns = col_n, index = index_n)
top5 = top5.fillna(0)

for i in range(len(top5)):
    top = abs_df[index_n[i]].nlargest(n=5)
    top5.iloc[i,:] = top.index
top5
```

```python
top5.to_csv('../data/top5_cat.csv')
df = pd.read_csv('../data/top5_cat.csv', index_col = 0)
df
```
